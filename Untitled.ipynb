```python
import pandas as pd
import numpy as np


file_path = 'Automobile_data.csv'
df = pd.read_csv(file_path)

df.replace('?', np.nan, inplace=True)


numeric_columns = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')


df.dropna(subset=['price'], inplace=True)

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())


categorical_columns = df.select_dtypes(include='object').columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])


df.reset_index(drop=True, inplace=True)


print("Cleaned Data Info:")
print(df.info())

print("\nCleaned Dataset Preview:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = 'Automobile_data.csv'
df = pd.read_csv(file_path)


df.replace('?', pd.NA, inplace=True)
numeric_columns = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df.dropna(subset=['price'], inplace=True)
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
categorical_columns = df.select_dtypes(include='object').columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])


print("\nSummary Statistics:")
print(df.describe(include='all'))


num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(bins=15, figsize=(15, 10), color='skyblue')
plt.suptitle('Histograms of Numeric Features')
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(x='body-style', y='price', data=df)
plt.title('Price by Body Style')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.title('Price by Drive Wheels')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = 'Automobile_data.csv'
df = pd.read_csv(file_path)
df.replace('?', pd.NA, inplace=True)
num_cols = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
df.dropna(subset=['price'], inplace=True)
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

plt.figure(figsize=(8, 6))
avg_price_dw = df.groupby('drive-wheels')['price'].mean().sort_values()
sns.barplot(x=avg_price_dw.index, y=avg_price_dw.values, palette='coolwarm')
plt.title('Average Price by Drive Wheels')
plt.ylabel('Average Price')
plt.xlabel('Drive Wheels')
plt.show()


plt.figure(figsize=(8, 6))
cross_tab = pd.crosstab(df['body-style'], df['drive-wheels'])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Body Style vs Drive Wheels')
plt.ylabel('Body Style')
plt.xlabel('Drive Wheels')
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x='fuel-type', y='horsepower', data=df, palette='Set2')
plt.title('Horsepower by Fuel Type')
plt.show()

plt.figure(figsize=(8, 6))
df_sorted = df.sort_values(by='highway-mpg')
sns.lineplot(x='highway-mpg', y='price', data=df_sorted, marker='o')
plt.title('Price vs Highway MPG')
plt.xlabel('Highway MPG')
plt.ylabel('Price')
plt.show()
```

```python

import pandas as pd


conclusions = {}


correlations = df.corr(numeric_only=True)['price'].sort_values(ascending=False)
top_positive_corr = correlations[1:4]
top_negative_corr = correlations[-3:]

conclusions['Top Positive Correlations with Price'] = top_positive_corr.to_dict()
conclusions['Top Negative Correlations with Price'] = top_negative_corr.to_dict()


avg_price_fuel = df.groupby('fuel-type')['price'].mean().to_dict()
conclusions['Average Price by Fuel Type'] = avg_price_fuel


common_bodies = df['body-style'].value_counts().head(3).index
avg_price_body = df[df['body-style'].isin(common_bodies)].groupby('body-style')['price'].mean().to_dict()
conclusions['Average Price of Most Common Body Styles'] = avg_price_body

import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(conclusions)
```
